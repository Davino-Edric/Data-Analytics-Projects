```python
x = x.drop(columns=['B','CHAS','RAD'])
```

```python
def minmax(input):
    minim = input.min()
    maxim = input.min()
    scaler = (input - minim) / (maxim - minim)
    return scaler, minim, maxim
```

```python
def strat(x,y,training = 0.8):
    np.random.seed(42)
    yidx = {}
    for i, label in enumerate(y):
        if label not in yidx:
            yidx[label] = [i]
        else:
            yidx[label].append(i)
            
    tr = []
    te = []
    
    for i,label in yidx.items():
        shuffle = np.random.permutation(label)
        train = max(1, int(len(shuffle) * training))
        tr.extend(shuffle[:train])
        te.extend(shuffle[train:])
        
    xtr = x.iloc[tr].reset_index(drop=True)
    xte = x.iloc[te].reset_index(drop=True)
    ytr = y.iloc[tr].reset_index(drop=True)
    yte = y.iloc[te].reset_index(drop=True)
    
    return xtr,xte,ytr,yte
```

```python
xtr,xte,ytr,yte = strat(x,target,training=0.7)
```

```python
print(xtr.shape)
print(xte.shape)
print(ytr.shape)
print(yte.shape)
```

Preprocessing: 
1. Nan diganti mean
2. Outlier diganti mean
3. Dummy variable bisa di cek di dummy_df
4. stratified split, normalisasi
5. ada x_num, x_cat, sama x, yang di strat buat regresi yang x
6. x_num, x_cat dipakek buat cari tau variabel apa aja yg  di drop (drop variabel B, CHAS)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(xtr,ytr)
```

```python
ypred = model.predict(xte)
```

```python
mse = mean_squared_error(yte, ypred)  # Mean Squared Error
r2 = r2_score(yte, ypred)  # RÂ² Score

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(F'rmse = {np.sqrt(mse):.2f}')
print(f"RÂ² Score: {r2:.2f}")
```

```python
import joblib

# Save the trained model to a file
joblib.dump(model, "boston_housing_model.pkl")
```
