```python
x = df.drop(columns='MEDV')
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split 

xtr, xte, ytr, yte = train_test_split(x,y, stratify=y, test_size=0.2, random_state=42)
```

```python
print(x.shape)
print(y.shape)
```

```python
print(xtr.shape)
print(xte.shape)
print(ytr.shape)
print(yte.shape)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

xtr_bal, ytr_bal = smote.fit_resample(xtr,ytr)
```

```python
minim = xtr.min()
maxim = xtr.max()
ranger = (maxim - minim)
xtrs = (xtr_bal- minim) / ranger
xtes = (xte- minim) / ranger
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(xtrs,ytr_bal)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
ypred_knn = knn.predict(xtes)
knn_rep = classification_report(ypred_knn,yte)
knn_acc = accuracy_score(ypred_knn,yte)
print(knn_acc)
print(knn_rep)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=100, random_state=42,n_estimators=100)
rf.fit(xtrs,ytr_bal)
```

```python
ypred_rf = rf.predict(xtes)
rf_rep = classification_report(ypred_rf,yte)
rf_acc = accuracy_score(ypred_rf,yte)
print(rf_acc)
print(rf_rep)
```
