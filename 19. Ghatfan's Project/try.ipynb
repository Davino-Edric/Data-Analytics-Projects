```python
mapping = {}
diction = {}

for col in dff.select_dtypes('object'):
    unique = dff[col].unique()
    mapping = {label : idx for idx,label in enumerate(unique)}
    diction[col] = mapping
    dff[col] = dff[col].map(mapping)
    
dff.head()
```

```python
def minmaxscaler(input):
    minim = input.min()
    maxim = input.max()
    norm = (input - minim) / (maxim-minim)
    return norm, minim, maxim
```

```python
dff.info()
```

```python
dfn,minim,maxim = minmaxscaler(dff)
```

```python
dff.head()
```

```python
dfn.head()
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(dfn.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
dfn.drop(columns=['Channel', 'Location','CustomerOccupation'],inplace=True)
```

```python
dfn.head()
```

```python
dff.info()
```

```python
from sklearn.cluster import DBSCAN, KMeans

inertia = []
k_range = range(2,11)

for k in k_range:
    kmeans= KMeans(n_clusters=k,random_state=42,n_init=10)
    kmeans.fit(dfn)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
dff['KMeans_Cluster'] = kmeans.fit_predict(dfn)

# Visualizing KMeans Clusters (using two most important features)
sns.scatterplot(x=dff['TransactionAmount'], y=dff['AccountBalance'], hue=dff['KMeans_Cluster'], palette='viridis')
plt.title('K-Means Clustering')
plt.show()
```
