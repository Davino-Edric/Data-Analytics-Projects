```python
for col in ndf.columns:
    q1 = ndf[col].quantile(0.25)
    q3 = ndf[col].quantile(0.75)
    iqr = q3 - q1
    lwr = q1 - 1.5 * iqr
    upr = q3 + 1.5 * iqr
    
    zs = (ndf[col] - ndf[col].mean()) / ndf[col].std()
    odf_iqr = ndf.loc[(ndf[col] < lwr) | (ndf[col] >= upr)]
    odf_zs = ndf[np.abs(zs > 2)]
```

```python
categories = ["Coffee & Tea", "Breakfast", "Chicken & Fish", "Beef & Pork", "Snacks & Sides"]
```

```python
def plot_top5_by_category(df, category, metric):
    category_df = df[df['Category'] == category]
    top5 = category_df.nlargest(5, metric)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=metric, y='Item', data=top5, palette='viridis', hue='Item', dodge=False, legend=False)
    plt.title(f'Top 5 {metric} in {category}')
    plt.xlabel(metric)
    plt.ylabel('Menu Item')
    plt.legend([],[], frameon=False)
    plt.show()
```

```python
metrics = ['Serving Size Oz', 'Calories', 'Total Fat']
for category in categories:
    for metric in metrics:
        plot_top5_by_category(df_reg, category, metric)
```

```python
categories = ["Coffee & Tea", "Breakfast", "Chicken & Fish", "Beef & Pork", "Snacks & Sides"]

# Fitur yang akan divisualisasikan
features = ["Serving Size Oz", "Calories", "Total Fat"]

# Loop untuk setiap fitur
for feature in features:
    plt.figure(figsize=(15, 8))
    
    # DataFrame untuk menyimpan hasil
    top_menus = pd.DataFrame()
    
    # Loop untuk setiap kategori
    for cat in categories:
        # Filter data berdasarkan kategori
        df_cat = df_reg[df_reg["Category"] == cat]
        
        # Ambil 5 menu dengan nilai tertinggi untuk fitur yang sedang di-loop
        top5 = df_cat.nlargest(5, feature)[["Item", feature, "Category"]]
        
        # Gabungkan ke dataframe hasil
        top_menus = pd.concat([top_menus, top5])
    
    # Plot bar chart
    plt.figure(figsize=(12, 6))
    sns.barplot(x=feature, y="Item", hue="Category", data=top_menus, dodge=False)
    
    plt.title(f"Top 5 Menu dengan {feature} Tertinggi per Kategori")
    plt.xlabel(feature)
    plt.ylabel("Menu")
    plt.legend(title="Category")
    plt.show()
```
