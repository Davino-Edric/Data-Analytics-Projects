```python
major = df[df['Diabetes_Type'] == 0]
minor = df[df['Diabetes_Type'] == 1]
mid = df[df['Diabetes_Type'] == 2]
mid_size = len(mid)

minor_scaled = minor.sample(mid_size, replace=True)

major_scaled = major.sample(mid_size,replace=True)

dfs = pd.concat([minor_scaled,major_scaled,mid])

df = dfs.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
df['Diabetes_Type'].value_counts()
```

```python
y = df['Diabetes_Type']
x = df.drop(columns='Diabetes_Type')
```

```python
def stratified(x,y, train_size = 0.8):
    np.random.seed(42)
    
    y_idx = {}
    for i, label in enumerate(y):
        if label not in y_idx:
            y_idx[label] = [i]
        else:
            y_idx[label].append(i)
            
    train, test = [], []
    for label, idx in y_idx.items():
        shuffle = np.random.permutation(idx)
        split = max(1, int(len(shuffle) * (train_size)))
        train.extend(shuffle[:split])
        test.extend(shuffle[split:])
        
        x_train = x.iloc[train].reset_index(drop=True)
        x_test = x.iloc[test].reset_index(drop=True)
        y_train = y.iloc[train].reset_index(drop=True)
        y_test = y.iloc[test].reset_index(drop=True)
        
    return x_train,x_test,y_train,y_test
```

```python
xtr,xte,ytr,yte = stratified(x,y)
```

```python
print(xtr.shape,xte.shape,ytr.shape,yte.shape)
```

```python
feat_train = xtr.to_numpy()
feat_test = xte.to_numpy()
label_train = ytr.to_numpy()
label_test = yte.to_numpy()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(feat_train, label_train)
```

```python
ypred = model.predict(feat_test)
accuracy = np.mean(yte == ypred)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report

print(classification_report(label_test, ypred))
```
