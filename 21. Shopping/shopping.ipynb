```python
x = df.drop(columns=['Customer ID','Location','Season','Review Rating','Shipping Type','Preferred Payment Method','Item Purchased','Category','Payment Method'])
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(x.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
dfs = scale.fit_transform(x)
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1,11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(dfs)
    wcss.append(model.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(2,11):
    model = KMeans(n_clusters=k,random_state=42)
    cluster = model.fit_predict(dfs)
    score = silhouette_score(dfs,cluster)
    sil_scores.append(score)
    
plt.plot(range(2,11),sil_scores, marker='o')
plt.grid(True)
plt.show()
```
