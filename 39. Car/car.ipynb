```python
dfc['brand'] = dfc['CarName'].str.extract(r'^(\S+)\s')
dfc.drop(columns='CarName',inplace=True)
dfc['brand'].value_counts()
```

```python
map_brand = {
    'maxda': 'mazda',
    'vw': 'volkswagen',
    'porcshce': 'porsche',
    'Nissan': 'nissan',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'alfa-romero': 'alfa-romeo',
}

dfc['brand'] = dfc['brand'].replace(map_brand)

dfc['brand'].value_counts()
```

```python
for col in dfc.select_dtypes('object'):
    if col != 'brand':
        plt.figure(figsize=(12,6))
        sns.boxenplot(x=dfc[col], y=dfc['price'])
        plt.show()
    else:
        plt.figure(figsize=(15,8))
        sns.boxenplot(y=dfc['brand'],x=dfc['price'])
        plt.show()
```

```python
for col in dfc[['cylindernumber','horsepower','enginelocation','doornumber','boreratio','enginesize','drivewheel']]:
        plt.figure(figsize=(15,8))
        sns.boxenplot(y=dfc['brand'],x=dfc[col])
        plt.show()
```

```python
expy = dfc.groupby('brand')['price'].median()
```

```python
expy
```

```python
dfc['brand_group'] = dfc['brand'].apply(lambda x: 0 if expy[x] < 10000 else (1 if (expy[x] < 20000) and (expy[x]> 10000) else 2))
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x=dfc['brand_group'],y=dfc['price'], hue=dfc['price_group'])
plt.legend(['Murah','Sedang','Mahal','Luxury'])
plt.show()
```
