```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CarPrice_Assignment.csv')
df.head()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.drop(columns='car_ID',inplace=True)
```

```python
for col in df.columns:
    sns.boxplot(df[col])
    plt.show()
```

```python
for col in df.columns:
    fig, axes = plt.subplots(1,2, figsize=(12,5))
    
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Histogram {col}')
    
    sns.boxenplot(df[col], ax=axes[1])
    axes[1].set_title(f'Boxenplot {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
df['brand'] = df['CarName'].str.extract(r'^([A-Za-z0-9\-\_]+)s*')
```

```python
df['brand'].value_counts()
```

```python
df.isna().sum()
```

```python
map_brand = {
    'toyouta':'toyota',
    'vokswagen':'volkswagen',
    'vw':'volkswagen' ,
    'porcshce':'porsche' ,
    'Nissan':'nissan',
    'maxda' : 'mazda'
}

df['brand'] = df['brand'].replace(map_brand)
df['brand'].value_counts()
```

```python
df.head()
```

```python
brand_val = df.groupby('brand')['price'].mean().sort_values(ascending=True).index
brand_val
```

```python
brand_val_dicto = {label : idx for idx, label in enumerate(brand_val)}
brand_val_dicto
```

```python
df['brand'] = df['brand'].map(brand_val_dicto)
```

```python
dfc = df.copy(deep=True)
dicto = {}

for col in dfc.select_dtypes('object'):
    dfc[col],dicto[col] = pd.factorize(dfc[col])
    
dfc.head()
```
