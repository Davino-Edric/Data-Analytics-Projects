## Melakukan Mapping pada kolom sesuai dengan kebutuhan masing-masing serta menambahkan kolom baru, mutation dan spread level

```python
df['Risk'] = df['Risk'].str.strip().map({
    'Low':0,
    'Intermediate':1,
    'High':2
})
df['Gender'] = df['Gender'].str.strip().map({
    'M':0,
    'F':1
})
df['Recurred'] = df['Recurred'].str.strip().map({
    'No':0,
    'Yes':1
})
df['HasAdenopathy'] = df['Adenopathy'].str.strip().apply(lambda x: 0 if x == 'No' else 1) # 0 None, 1 = sisanya
df['Hx Radiotherapy'] = df['Hx Radiotherapy'].str.strip().map({
    'No':0,
    'Yes':1
})
df['Focality'] = df['Focality'].str.strip().map({
    'Uni-Focal':0,
    'Multi-Focal':1
})
df['IsEarly'] = df['Stage'].str.strip().apply(lambda x: 1 if x == 'I' else 0)
df['Stage'] = df['Stage'].str.strip().map({
    'I':0,
    'II':1,
    'III':2,
    'IVA':3,
    'IVB':4
})
df['N'] = df['N'].map({
    'N0':0,
    'N1':1
})
df['M'] = df['M'].map({
    'M0':0,
    'M1':1
})
df['T'] = df['T'].map({
    'T1':0,
    'T2':1,
    'T3':2,
    'T4':3
})
df['Adenopathy'] = df['Adenopathy'].map({label:idx for idx,label in enumerate(df['Adenopathy'].unique())})

df['Mutation'] = df['T'] + df['N'] + df['M']
df['Spread_level'] = df['Mutation'] * (df['Focality'] +1) + df['Stage']
df
```

- Kolom Mutation dibuat untuk mengsimplifikasi data mutasi T, N, M untuk mencari nilai tingkatan mutasi secara total
- Kolom Spread level adalah kolom untuk mengetahui penyebaran dari kanker tersebut, meskipun tidak 100% benar secara rumus, dengan data yang dimiliki, ini merupakan step terbaik untuk mengukur seberapa parah penyebaran kanker tersebut

## 3.1 Melakukan stratified split 

```python
x = df.drop(columns='Recurred').values
y = df['Recurred'].values
cols = df.drop(columns='Recurred').columns
cols
```

```python
def strat_split(x, y, test_size=0.2, stratify=None, random_state=42):
    stratify = stratify if stratify is not None else y
    np.random.seed(random_state)
    test_indices = []
    train_indices = []
    unique_class, y_indices = np.unique(stratify, return_inverse=True)
    for class_index in range(len(unique_class)):
        class_indices = np.where(y_indices == class_index)[0]
        np.random.shuffle(class_indices)
        n_test = int(len(class_indices) * test_size)
        test_indices.extend(class_indices[:n_test])
        train_indices.extend(class_indices[n_test:])

    test_indices = np.array(test_indices)
    train_indices = np.array(train_indices)
    x_train, x_test = x[train_indices], x[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]

    return x_train, x_test, y_train, y_test
```

## 3.2 Melakukan Normalisasi

```python
def normalisasi(train, test):
    min_val = np.min(train, axis=0)
    max_val = np.max(train, axis=0)
    denominator = max_val - min_val
    denominator = denominator.astype(float)
    denominator[denominator == 0] += 1e-10
    train = (train - min_val) / denominator
    test = (test - min_val) / denominator
    return train, test
```

## 3.3. Melakukan synthetic oversampling SMOTE

```python
def smote(x, y, k=None, random_state=42):
    np.random.seed(random_state)
    unique_classes = np.unique(y)
    class_counts = {class_label: np.sum(y == class_label) for class_label in unique_classes}
    max_class_counts = max(class_counts.values())
    x_resampled = x.tolist()
    y_resampled = y.tolist()
    for class_label in unique_classes:
        if class_counts[class_label] < max_class_counts:
            class_indices = np.where(y == class_label)[0]
            class_samples = x[class_indices].astype('float64')
            n_samples = max_class_counts - class_counts[class_label]
            for _ in range(n_samples):
                sample_index = np.random.choice(class_indices)
                sample = x[sample_index].astype('float64')
                distances = np.linalg.norm(class_samples - sample, axis=1)
                nearest_neighbour = np.argsort(distances)[1:k+1]
                neighbour_index = np.random.choice(nearest_neighbour)
                neighbour = class_samples[neighbour_index]
                diff = neighbour - sample
                synthetic_sample = diff + np.random.rand() * diff
                x_resampled.append(synthetic_sample)
                y_resampled.append(class_label)

    return np.array(x_resampled), np.array(y_resampled)
```

```python
x_train_resampled, y_train_resampled = smote(x_train_scaled, y_train, k=5)
```

# 3.4 Model KNN Manual

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def manhattan_distance(a, b):
    return np.sum(np.abs(a - b))

def minkowski_distance(a, b, p):
    return np.sum(np.abs(a - b) ** p) ** (1 / p)

class KNearestNeighbours:
    def __init__(self, n_neighbours=5, metric='euclidean', p=3):
        self.n_neighbours = n_neighbours
        self.metric = metric.strip().lower()
        self.p = p

    def fit(self, X, y, sequential=False,threshold_max=0.4,step=0.2,column_name=None):
        self.y_train = y
        self.X_train = X
        self.x_backup = X
        self.y_backup = y
        self.sequential = sequential
        self.thresh = threshold_max
        self.cols = column_name
        self.step = step
        self.best_feat = None
        
        if self.sequential is True:
            best_score = 0
            self.x_sub,self.x_val,self.y_sub,self.y_val = strat_split(self.X_train,self.y_train,stratify=y, random_state=42)
            
            dataset = pd.DataFrame(self.x_sub,columns=self.cols)
            dataset['target'] = self.y_sub
            validation = pd.DataFrame(self.x_val,columns=self.cols)
            validation['target'] = self.y_val
            
            corr = dataset.drop(columns='target').apply(lambda x: dataset['target'].corr(x))
            dataset = dataset.drop(columns='target')
            for i in np.round(np.arange(0.1,self.thresh,self.step), 2):
                print(i)
                self.selected = corr[corr > i].index.tolist()
                self.best_feat = self.selected
                self.x_sub = dataset[self.selected].values
                self.x_val = validation[self.selected].values
                
                self.X_train = self.x_sub
                self.y_train  = self.y_sub
                preds = self.predict(self.x_val)
                acc = np.mean(preds == self.y_val)
                if acc > best_score:
                    best_score = acc
                    self.best_feat = self.selected
                    print(f'Found best threshold {i} with accuracy of: {best_score}')
                    print(f'Selected Features: {self.best_feat}')
                    
            self.X_train = self.x_backup
            self.y_train = self.y_backup
        
        if self.best_feat is None:
            self.best_feat = self.cols
            print('Best feature is all feature.')
            
    def predict(self, X):
        if self.sequential is True:
            print(self.best_feat)
            X = pd.DataFrame(X,columns=self.best_feat).values
        predictions = [self._predict(x) for x in X]
        return np.array(predictions)
    
    def _predict(self, x):
        distances = [self._distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.n_neighbours]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        most_common = np.bincount(k_nearest_labels).argmax()
        return most_common
    
    def _distance(self, a, b):
        if self.metric == 'euclidean':
            return euclidean_distance(a, b)
        elif self.metric == 'manhattan':
            return manhattan_distance(a, b)
        elif self.metric == 'minkowski':
            return minkowski_distance(a, b, self.p)
        else:
            raise ValueError(f"Unknown metric: {self.metric}")
```

## **4.2 Classification Report**

```python
def precision_recall_f1_support(y_true, y_pred, average=None):
    labels = np.unique(np.concatenate((y_true, y_pred)))
    report = {}
    
    for label in labels:
        tp = sum((y_true[i] == label) and (y_pred[i] == label) for i in range(len(y_true)))
        fp = sum((y_true[i] != label) and (y_pred[i] == label) for i in range(len(y_true)))
        fn = sum((y_true[i] == label) and (y_pred[i] != label) for i in range(len(y_true)))
        tn = sum((y_true[i] != label) and (y_pred[i] != label) for i in range(len(y_true)))

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        support = sum(y == label for y in y_true)

        report[label] = {
            'precision': precision,
            'recall': recall,
            'f1-score': f1_score,
            'support': support
        }

    if average == 'macro':
        avg_precision = sum(report[label]['precision'] for label in labels) / len(labels)
        avg_recall = sum(report[label]['recall'] for label in labels) / len(labels)
        avg_f1_score = sum(report[label]['f1-score'] for label in labels) / len(labels)
        avg_support = sum(report[label]['support'] for label in labels)
        
        report['macro avg'] = {
            'precision': avg_precision,
            'recall': avg_recall,
            'f1-score': avg_f1_score,
            'support': avg_support
        }
    
    return report
```

```python
def print_classification_report(report):
    print(f"{'Label':<10} {'Precision':<10} {'Recall':<10} {'F1-Score':<10} {'Support':<10}")
    print("="*50)
    for label, metrics in report.items():
        print(f"{label:<10} {metrics['precision']:<10.2f} {metrics['recall']:<10.2f} {metrics['f1-score']:<10.2f} {metrics['support']:<10}")
```

```python
report_knn = precision_recall_f1_support(y_testb, y_pred, average='macro')
print_classification_report(report_knn)
```

## **4.3 Confusion Matrix**

```python
def confusion_matrix(y_test, y_pred):
    tp = fp = tn = fn = 0
    for true, pred in zip(y_test, y_pred):
        if true == 1 and pred == 1:
            tp += 1
        elif true == 0 and pred == 1:
            fp += 1
        elif true == 0 and pred == 0:
            tn += 1
        elif true == 1 and pred == 0:
            fn += 1
    
    confusion_matrix = np.array([[tn, fp], [fn, tp]])

    fig, ax = plt.subplots()
    ax.matshow(confusion_matrix, cmap='coolwarm', alpha=0.6)
    for i in range(confusion_matrix.shape[0]):
        for j in range(confusion_matrix.shape[1]):
            ax.text(x=j, y=i, s=confusion_matrix[i, j], va='center', ha='center')

    plt.xlabel('Prediksi')
    plt.ylabel('Aktual')
    plt.title('Confusion Matrix')
    plt.xticks([0, 1], ['Sembuh', 'Kambuh'])
    plt.yticks([0, 1], ['Sembuh', 'Kambuh'])
    plt.show()
```

```python
def normalkan(train, test):
    min_val = np.min(train, axis=0)
    max_val = np.max(train, axis=0)
    denominator = max_val - min_val
    denominator = denominator.astype(float)
    denominator[denominator == 0] += 1e-10
    train = (train - min_val) / denominator
    test = (test - min_val) / denominator
    return train, test, min_val,denominator
```
