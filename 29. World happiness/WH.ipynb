# Soal 2

## IQR & Z Score

```python
odf_iqr = {}
for col in df_num.columns:
    q1 = df_num[col].quantile(0.25)
    q3 = df_num[col].quantile(0.75)
    iqr = q3 - q1 
    upr = q3 + 1.5 * iqr
    lwr = q1 - 1.5 * iqr
    
    zs = (df_num[col] - df_num[col].mean()) / df_num[col].std()
    
    odf_iqr = df_num.loc[(df_num[col] < lwr) | (df_num[col] > upr)]
    odf_zs = df_num[np.abs(zs) > 3]
```

## IQR


```python
odf_iqr
```

## Z score

```python
odf_zs
```

## Covariance correlation

```python
cov_dict = {}
for col in df_num.columns:
    cov_dict[col] = np.cov(df_num[col],df_num['Score'], bias=True)[0][1]
    
cov_dict
```
