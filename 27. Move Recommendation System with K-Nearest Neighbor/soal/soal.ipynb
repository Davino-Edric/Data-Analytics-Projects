```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv_file.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import ast
def nilai_pertama(kolom):
    if kolom:
        try:
            kolom = kolom.replace('\\"', "'")
            kolom_list = ast.literal_eval(kolom)
            if kolom_list:
                return kolom_list[0]['name']
        except (SyntaxError, ValueError):
            return None
        return None
    
df['genre'] = df['genres'].apply(nilai_pertama)
df['production_company'] = df['production_companies'].apply(nilai_pertama)
df.drop(columns=['genres', 'production_companies'], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
title = df['original_title']
df = df[['budget', 'genre', 'production_company','popularity','revenue','vote_average','vote_count']]
df.head()
```

```python
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
```

```python
df.dtypes
```

```python
cat_col = df.select_dtypes(include='object')
mapping = {}
dictio = {}
for col in cat_col:
    unique = cat_col[col].unique()
    mapping = {label : idx for idx,label in enumerate(unique)}
    dictio[col] = mapping
    df[col] = df[col].map(mapping)
```

```python
dictio
```

```python
df.head()
```

```python
dff = df[['budget', 'production_company','popularity','revenue','vote_average','vote_count']]
min_value = dff.min(axis=0)
max_value = dff.max(axis=0)
range_value = max_value - min_value
range_value[range_value == 0] = 1
df_scaled = (dff - min_value)/range_value
```

```python
for col in df_scaled.columns:
    sns.boxenplot(df[col])
    plt.show()
```

```python
for col in df_scaled.columns:
    q1 = df_scaled[col].quantile(0.25)
    q3 = df_scaled[col].quantile(0.75)
    iqr = q3 - q1 
    print(q1,q3,iqr)
    lwr = q1 - 1.5 * iqr
    upr = q3 + 1.5 * iqr
    df_scaled = df_scaled[(df_scaled[col] >= lwr) & (df_scaled[col] <= upr)]
    
```

```python
df_scaled['title'] = title.reset_index(drop=True)
df_scaled['genre'] = df['genre'].reset_index(drop=True)
df_scaled.head()
```

```python
df_scaled.to_csv('df_scaled.csv')
```
