```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('game_of_thrones_test.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.fillna(0,inplace=True)
df
```

```python
df.drop(columns='S.No',inplace=True)
```

```python
for i in df.select_dtypes('object'):
    df[i], unique = pd.factorize(values=df[i])
```

```python
df
```

```python
sns.histplot(df['popularity'])
plt.show()
```

```python
df['log_pop'] = np.log1p(df['popularity'])
sns.histplot(df['log_pop'])
plt.show()
```

```python
df['log_pop1'] = np.log1p(df['log_pop'])
sns.histplot(df['log_pop1'])
plt.show()
```

```python
df['pop_class'] = pd.cut(df['log_pop1'],bins=[0,0.18,0.35,0.6], labels=[0,1,2])
df.drop(columns=['log_pop','log_pop1'],inplace=True)
df['pop_class'].value_counts()
```

```python
plt.figure(figsize=(17,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```
