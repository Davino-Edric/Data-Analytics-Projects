```python
for col in df[['chol','trestbps']]:
    q1 = dfc[col].quantile(0.25)
    q3 = dfc[col].quantile(0.75)
    iqr = q3 - q1
    lwr = q1 - 1.5 * iqr
    upr = q3 + 1.5 * iqr
    
    dfo = dfc.loc[(dfc[col] < lwr) | (dfc[col] > upr)]
```

```python
x = df.drop(columns='target')
y = df['target']
```

```python
def strat(x,y, train_size = 0.8):
    yidx = {}
    for i,label in enumerate(y):
        if label not in yidx:
            yidx[label] = [i]
        else:
            yidx[label].append(i)
            
    train_idx = []
    test_idx = []
    
    for i,label in yidx.items():
        shuffle = np.random.permutation(label)
        set_idx = max(1, int(len(shuffle) * train_size))
        train_idx.extend(shuffle[:set_idx])
        test_idx.extend(shuffle[set_idx:])
        
    xtr = x.iloc[train_idx].reset_index(drop=True)
    xte = x.iloc[test_idx].reset_index(drop=True)
    ytr = y.iloc[train_idx].reset_index(drop=True)
    yte = y.iloc[test_idx].reset_index(drop=True)
    
    return xtr,xte,ytr,yte
```

```python
xtr,xte,ytr,yte = strat(x,y)
```
