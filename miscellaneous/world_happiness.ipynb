```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2019.csv')
df.head()
```

```python
column_num = df.select_dtypes(include='number')
column_num.drop(columns='Overall rank', inplace=True)
```

```python
column_num.head()
```

# Soal 1

## Mean

```python
mean_val = column_num.mean()
mean_val
```

## Median

```python
med_val = column_num.median()
med_val
```

## Modus

```python
modus_val = column_num.mode().iloc[0]
modus_val
```

## Range

```python
min_val = column_num.min()
max_val = column_num.max()
range_val = max_val - min_val
range_val
```

## Q1

```python
q1 = column_num.quantile(0.25)
q1
```

## Q3

```python
q3 = column_num.quantile(0.75)
q3
```

## IQR

```python
iqr = q3 - q1
iqr
```

## Variance

```python
varian = column_num.var()
varian
```

## Standard Deviasi

```python
std_val = column_num.std()
std_val
```

## Z-score

```python
z_score = (column_num - mean_val) / std_val
z_score
```

# Soal 2

## Outlier dengan IQR method

```python
for col in column_num.columns:
    q1 = column_num[col].quantile(0.25)
    q3 = column_num[col].quantile(0.75)
    iqr = q3 - q1
    lwr = q1 - 1.5 * iqr
    upr = q3 + 1.5 * iqr
    outlier_iqr = (column_num[(column_num[col] < lwr) | (column_num[col] > upr)])
```

```python
outlier_iqr
```

## Outlier dengan Z-score method

```python
for col in column_num.columns:
    mean_val = column_num[col].mean()
    std_val = column_num[col].std()
    z_scores = (column_num[col] - mean_val) / std_val
    outliers_z = column_num[np.abs(z_scores) > 3]
```

```python
outliers_z
```
